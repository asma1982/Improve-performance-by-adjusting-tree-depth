##  Evaluation using cross validation


```python
## By KFold
```

```python

X = BC[['Adress','Age','FBS', 'Urea','ALB', 'TCa', 'GPT','ALP','CA15','CEA','WBC','RBC','PLT','ESR','LDH']].values
y = BC['CLASS'].values
```

```python
from sklearn.model_selection import train_test_split 

## spilt
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,stratify=y,random_state=101) 
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

D_classifier= DecisionTreeClassifier(random_state=0) 
```

```python

from sklearn.metrics import mean_squared_error as mse

max_depths = range(1, 30)
training_error = []
for max_depth in max_depths:
    model_1 = DecisionTreeClassifier(max_depth=max_depth)
    model_1.fit(X, y)
    training_error.append(mse(y_train, model_1.predict(X_train)))
    
testing_error = []
for max_depth in max_depths:
    model_2 = DecisionTreeClassifier(max_depth=max_depth)
    model_2.fit(X_train, y_train)
    testing_error.append(mse(y_test, model_2.predict(X_test)))

plt.plot(max_depths, training_error, color='blue', label='Training error')
plt.plot(max_depths, testing_error, color='green', label='Testing error')
plt.xlabel('Tree depth')
plt.axvline(x=3, color='orange', linestyle='--')
plt.annotate('optimum = 3', xy=(4.5, 3.10), color='red')
plt.ylabel('Mean squared error')
plt.title('Hyperparameter Tuning', size=15)
plt.legend()
plt.savefig('error.png')
```
